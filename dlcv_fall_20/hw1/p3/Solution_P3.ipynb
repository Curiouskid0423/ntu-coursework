# Problem 3: Image Filtering

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.ndimage import gaussian_filter
```

```python
lena = cv2.imread("../lena.png", 0)

plt.rcParams["figure.figsize"] = (10,5)
plt.title("Lena (Original Pic)")
plt.imshow(lena)
plt.show()
```

```python
lena
```

```python
gf_lena = gaussian_filter(lena, sigma =  1 / (2 * np.log(2)))
```

```python

plt.rcParams["figure.figsize"] = (10,5)
plt.title("Lena (GF Pic)")
plt.imshow(gf_lena)
plt.show()
```

```python
gf_lena
```

```python
gf_lena_2 = gaussian_filter(lena, sigma =  4)
plt.rcParams["figure.figsize"] = (10,5)
plt.title("Lena (GF Pic)")
plt.imshow(gf_lena_2)
plt.show()
```

### Part 1:  
- The result after applying Gaussian Filter with a miniscule sigma is insignificant to naked eyes. However, when applied a filter with higher sigma, the effect is that the image will be smoothened, which is the intention of a Gaussian Filter
